# 1) Write a python program to display all the header tags from wikipedia.org and make data frame.
URL : https://en.wikipedia.org/wiki/Main_Page

```python
wiki_page = requests.get('https://en.wikipedia.org/wiki/Main_Page') #Assign a variable to request the URL for response
wiki_page
```

```python
wiki_soup= BeautifulSoup(wiki_page.content) #Assign a variable to extract content using BeautifulSoup
wiki_soup
```

```python
wiki_header= [] #Create an empty list(container) #for loop to search for header tags #append the content to the empty list.

for i in wiki_soup.find_all(["h1", "h2", "h3", "h4", "h5", "h6"]):
    wiki_header.append(i.text.strip())

wiki_header
```

```python
wiki_df = pd.DataFrame({'Title':wiki_header}) #Assign a variable to generate pandas dataframe
wiki_df
```

# 2) Write a python program to display list of respected former presidents of India(i.e. Name , Term of office)
from https://presidentofindia.nic.in/former-presidents.htm and make data frame.

```python
pres_page = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
pres_page
```

```python
pres_soup = BeautifulSoup(pres_page.content)
pres_soup
```

```python
pres_name = []

for i in pres_soup.find_all("div",class_="presidentListing"):
    name = i.text.strip().split("(")[0].strip()
    pres_name.append(name)
pres_name
```

```python
pres_term = []

for i in pres_soup.find_all("div",class_="presidentListing"):
    term = i.text.strip().split(":")[1].split("\n")[0].strip()
    pres_term.append(term)
    
pres_term
```

```python
pres_image = []

for i in pres_soup.find_all("img", class_="alignLeft"):
    image = i.get("src")
    pres_image.append(image)

pres_image
```

```python
pres_df = pd.DataFrame({"Name": pres_name, "Term of Office": pres_term})

pd.set_option('display.max_colwidth', None)

pres_df
```

#### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
odi_teams = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
odi_teams
```

```python
soup_odi_team = BeautifulSoup(odi_teams.content)
soup_odi_team
```

```python
country = []
for i in soup_odi_team.find_all("span",class_="u-hide-phablet"):
    name = i.text
    country.append(name)
country = country[:10]

country
```

```python
matches = []
top_team = soup_odi_team.find("td", class_="rankings-block__banner--matches").text

for i in soup_odi_team.find_all("td",class_="table-body__cell u-center-text"):
    matches.append(i.text)
    
matches = matches[:18:2]

matches.insert(0, top_team)

matches
```

```python
points = []
top_point = soup_odi_team.find("td",class_="rankings-block__banner--points").text

for i in soup_odi_team.find_all("td",class_="table-body__cell u-center-text"):
    points.append(i.text)
    
points = points[1:18:2]

points.insert(0,top_point)
points
```

```python
ratings= []
top_rating = soup_odi_team.find('td',class_="rankings-block__banner--rating u-text-right").text.strip()

for i in soup_odi_team.find_all("td",class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
    
ratings= ratings[:9:]
ratings.insert(0,top_rating)
ratings
```

```python
odi_team_df= pd.DataFrame({"Country":country,"Matches":matches, "Points":points,"Ratings":ratings})

odi_team_df
```

# 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.

```python
ai_page= requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup_ai = BeautifulSoup(ai_page.content)
article=[]
date=[]
url=[]
author=[]

for i in soup_ai.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    article.append(i.text)
for i in soup_ai.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    date.append(i.text)
for i in soup_ai.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    url.append(i.get("href"))
for i in soup_ai.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    author.append(i.text)


article_df = pd.DataFrame({"Title":article, "Author":author, "Published": date, "Link": url})

article_df
```

# 7) Write a python program to scrape mentioned details from dineout and make data frame

```python
dineout_page= requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")

dineout_soup= BeautifulSoup(dineout_page.content)

do_names = []
do_loc = []
do_cuisine = []
do_price = []
do_rating = []
do_img=[]

for i in dineout_soup.find_all('a',class_="restnt-name ellipsis"):
    do_names.append(i.text)

for i in dineout_soup.find_all('div',class_="restnt-loc ellipsis"):
    do_loc.append(i.text)

for i in dineout_soup.find_all('span',class_="double-line-ellipsis"):
    do_cuisine.append(i.text.split("|")[1].strip())

for i in dineout_soup.find_all('span',class_="double-line-ellipsis"):
    do_price.append(i.text.split("|")[0].strip())


for i in dineout_soup.find_all("div",class_="restnt-rating rating-4"):
    do_rating.append(i.text)

for i in dineout_soup.find_all('img',class_="no-img"):
    do_img.append(i.get("data-src"))


dineout_df = pd.DataFrame({"Restaurant Name":do_names, "Cost": do_price,"Rating": do_rating, "Cuisine": do_cuisine, "Location": do_loc, "Images":do_img})

pd.set_option('display.max_colwidth', None)
dineout_df
```
